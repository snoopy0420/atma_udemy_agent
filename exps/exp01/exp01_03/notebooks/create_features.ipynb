# 特徴量生成

```python
def run_blocks(feature_blocks):
    print('start run blocks...')
    with Timer(prefix='run all'):
        for block in feature_blocks:
            with Timer(prefix=f'  - {block.__class__.__name__}'):
                feature = block.create_feature()
                print(f'    shape: {feature.shape}')
```

```python
os.makedirs(DIR_FEATURE, exist_ok=True)

feature_blocks = [
    Key(use_cache=False, save_cache=True),
    Target(use_cache=False, save_cache=True),
    CategoryFeature(use_cache=False, save_cache=True),
    UdemyActivityFeature(use_cache=False, save_cache=True),
    UdemyActivityDetailFeature(use_cache=False, save_cache=True),
    OvertimeWorkByMonthFeature(use_cache=False, save_cache=True),
    OvertimeTimeSeriesFeature(use_cache=False, save_cache=True),
    PositionHistoryFeature(use_cache=False, save_cache=True),
    DxFeature(use_cache=False, save_cache=True),
    HrFeature(use_cache=False, save_cache=True),
    CareerFeature(use_cache=False, save_cache=True),
]

run_blocks(feature_blocks)
```
